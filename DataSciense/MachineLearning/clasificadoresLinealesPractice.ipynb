```python
import pandas as pd
import numpy as np
import sklearn.model_selection as train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score, confusion_matrix 
```

DataSet que mide el nivel de morosidad de una persona en función de sus circunstancias

```python
data = pd.read_csv('./DataExampleCSV/Default.csv', sep="\t")
```

```python
data.head()
```

```python
data.default.value_counts()
```

```python
data.student.value_counts()
```

Convierto a Dummys las variables categóricas

```python
student_dummy = pd.get_dummies(data.student, prefix='student', drop_first=True, dtype=int)
student_dummy.head(3)
```

Concateno las tablas

```python
data = pd.concat([data, student_dummy], axis=1)  
```

```python
data.head(3)
```

Matriz de Features

```python
X= data[['student_Yes', 'balance', 'income']]
y = data.default
```

```python
x_train, x_test, y_train, y_test = train_test_split.train_test_split(X, y, stratify=y, random_state=42)
```

```python
data.default.value_counts("Yes")/data.shape[0]
```

```python
logistic_regression = LogisticRegression(penalty=None) # penalty implica que no tendrá regularización 
logistic_regression.fit(x_train, y_train)
```

```python
y_test_pred = logistic_regression.predict(x_test)
y_test_pred
```

```python
y_test_pred_proba = logistic_regression.predict_proba(x_test)
y_test_pred_proba
```

```python
accuracy_score(y_test, y_test_pred) # Nivel de confianza de predicción
```

```python
confusion_matrix(y_test, y_test_pred)
```
